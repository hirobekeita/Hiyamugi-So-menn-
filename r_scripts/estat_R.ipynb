```r
#Labor force survey basic tabulation
##############initial setup################
pacman::p_load(tidyverse, magrittr, stringr, estatapi)
############################################
```

```r
AppID <- "5cbc73bc48a1b7ded51b3efb188b32315ae717f8"
```

```r
ResultData <- estat_getStatsList(appId = AppID, searchWord = "労働力")
#confirm
head(ResultData)
```

```r
#############################################################
###statsDataID=0002060002:就業状態，年齢階級別15歳以上人口###
#############################################################

MetaInfo <- estat_getMetaInfo(appId = AppID, statsDataId = "0002060002")
#confirm
head(MetaInfo)
```

```r
GetStatData <- estat_getStatsData(appId = AppID,
                                  statsDataId = "0002060002",
                                  limit = 9999999)
head(GetStatData)
```

```r
GetStatData$cat04_code
```

```r
#####年齢は5歳階級のみを抽出#####
d <- dplyr::filter(GetStatData, cat04_code %in% c("00", "27", "02", "05","07","08","10","11","13","14","16","17","19","28","30","31","32"))
head(d)
tail(d)
```

```r
#####完全失業者のみを抽出#####
d2 <- dplyr::filter(d, cat03_code=="08")
head(d2)
tail(d2)
```

```r
#####データ整形，リネームを行う#####
df <- d2 %>% dplyr::select("産業","性別", "就業状態", "年齢階級", "time_code","value", "unit") %>% dplyr::rename(industry="産業",gender="性別", condition="就業状態", age="年齢階級")
year <- str_sub(df$time_code, start=1, end=4)
#head(year)
month <- str_sub(df$time_code, start=9, end=10)
#head(month)
ym <- paste(year, month, sep="-")
```

```r
###定義したymをtime_codeとしてdfをmutate
df <- dplyr::mutate(df, time_code = ym)
head(df)
```

```r
write.csv(df, "labor_survey_unemployed.csv", fileEncoding="UTF-8", row.names=FALSE)
```

```r
###########################################################################
###statsDataID=0002060004:年齢階級別労働力人口比率，就業率及び完全失業率###
###########################################################################

ResultData2 <- estat_getStatsList(appId = AppID, searchWord = "失業率")
#confirm
head(ResultData2)
```

```r
MetaInfo <- estat_getMetaInfo(appId = AppID, statsDataId = "0002060004")
#confirm
MetaInfo
```

```r
GetStatData2 <- estat_getStatsData(appId = AppID,
                                   statsDataId = "0002060004",
                                   limit = 9999999)
head(GetStatData2)
```

```r
#####年齢は5歳階級のみを抽出#####
d3 <- dplyr::filter(GetStatData2, cat04_code %in% c("00", "27", "02", "05","07","08","10","11","13","14","16","17","19","28","30","31","32"))
#####完全失業者のみを抽出#####
d4 <- dplyr::filter(d3, cat03_code=="08")
#####データ整形，リネームを行う#####
df2 <- d4 %>% dplyr::select("産業","性別", "就業状態", "年齢階級", "time_code","value", "unit") %>% dplyr::rename(industry="産業",gender="性別", condition="就業状態", age="年齢階級")
year <- str_sub(df2$time_code, start=1, end=4)
#head(year)
month <- str_sub(df2$time_code, start=9, end=10)
#head(month)
ym <- paste(year, month, sep="-")
#ym
df2 <- dplyr::mutate(df2, time_code = ym)
head(df2)
write.csv(df2, "labor_survey_unemployed2.csv", fileEncoding="UTF-8", row.names=FALSE)
```

```r
dat <- dplyr::inner_join(df , df2 ,by=c("gender", "age", "time_code")) %>% dplyr::select("industry.x","gender", "condition.x", "age", "time_code", "value.x","value.y") %>% dplyr::rename(industry="industry.x", condition="condition.x", volume="value.x", ratio="value.y")
head(dat)
#dfのunitは単位が万人なので，予め10000を乗算する
dat2 <- dplyr::mutate(dat, volume = volume*10000)
#head(dat2)
```
